## Correction du 28




*indice: Médiatrice*

On cherche donc à vérifier qu'il y a équidistance entre les points proposés et les extrémités du segment.   
Si A appartient à la médiatrice de $[IJ]$ alors $AI=AJ$.  
  
$$\begin{array}
AI=\sqrt{(x_I-x_A)^2+(y_I-y_A)^2} & AJ=\sqrt{(x_J-x_A)^2+(y_J-y_A)^2}\\ 
AI= \sqrt{(-1-1)^2+(2-1)^2} & AJ=\sqrt{(2-1)^2+(3-1)^2}\\
AI=\sqrt{4+1} & AJ = \sqrt{1+4}\\
AI = \sqrt{5} & AJ=\sqrt{5}
\end{array}$$  

Donc A appartient à la médiatrice de $[IJ]$.  
  
  B appartient à la médiatrice de $[IJ]$ ssi $BI=BJ$.  
  $$\begin{array}
BI=\sqrt{(x_I-x_B)^2+(y_I-y_B)^2} & BJ=\sqrt{(x_J-x_B)^2+(y_J-y_B)^2}\\ 
BI= \sqrt{(-1+0,5)^2+(5-2)^2} & BJ=\sqrt{(2+0,5)^2+(3-5)^2}\\
BI=\sqrt{0,25+9} & BJ = \sqrt{6,25+4}\\
BI = \sqrt{9,25} & AJ=\sqrt{10,25}
\end{array}$$  
Donc B n'appartient pas à la médiatrice de $[IJ]$.  
  
  C appartient à la médiatrice de $[IJ]$ ssi $CI=CJ$.  
  $$\begin{array}
CI=\sqrt{(x_I-x_C)^2+(y_I-y_C)^2} & CJ=\sqrt{(x_J-x_C)^2+(y_J-y_C)^2}\\ 
CI= \sqrt{(-1+2)^2+(2-2)^2} & CJ=\sqrt{(2+2)^2+(3-2)^2}\\
CI=\sqrt{1+0} & CJ = \sqrt{1+4}\\
CI = \sqrt{1} & CJ=\sqrt{5}
\end{array}$$    
Donc C n'appartient pas à lamédiatrice de $[IJ]$.

## correction du 36
  


1) distance $CA$  
Dans le triangle CAB rrectangle en A, on applique le théorème de Pythagore:  
$$\begin{array} \\ 
CB^2&=&AB^2+CA^2\\
 CA^2&=&CB^2-AB^2\\
CA^2&=&10^2-4^2\\
CA^2&=&100-16\\
CA^2&=&84\\
CA&=&\sqrt{84}
\end{array}$$

$$\begin{array} \\ 
CI^2&=&AI^2+CA^2\\
CI^2&=&2^2+\sqrt{84}^2\\
CI^2&=&4+84\\
CI^2&=&88\\
CI&=&\sqrt{88}
\end{array}$$
