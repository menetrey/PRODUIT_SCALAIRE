##  Histoire des Mathématiques  
La notion de vecteur était implicite en mécanique depuis Galilée mais a mis longtemps à prendre sa forme actuelle.   
    
  
  La notion de vecteur apparaît chez Leibniz au cours de ses recherches sur l'élaboration d'un calcul des variations.  

Le $XIX^e$ siècle voit apparaître la notion de produit scalaire en mathématiques et la notion de travail en physique. Ces notions offrent l'avantage de combiner vision géométrique et calcul.  
Les cercles font partie des plus vieux objets mathématiques. La caractérisation du cercle de diamètre $[AB]$ comme ensemble des points M tels que AMB soit un triangle rectangle en M semble remonter à Thalès.  
Mais ce n'est qu'au $XVII^e$ siècle que Descartes élabore la méthode des coordonnées et écrit l'équation d'un cercle en repère orthonormé.  



## I) Rappels sur les vecteurs
On définit les points A, B, C et D quatre points du plan. 
  
  ### 1) Egalité de deux vecteurs
  $\overrightarrow{AB}=\overrightarrow{CD} \iff ABDC$ est un parallélogramme.  
  
  ### 2) Relation de Chasles
  $\overrightarrow{AC}=\overrightarrow{AB}+\overrightarrow{BC}$
  
  ### 3) Vecteurs opposés
  $-\overrightarrow{AB}=\overrightarrow{BA}$
  
  ### 4) Vecteur $k\overrightarrow{u}$
  Soit $\overrightarrow{u}$ un vecteur et k un scalaire (réel) .  
  
  La longueur de $\overrightarrow{u}$ est multiplié par |k|:  
• Si $k > 0$ son sens est inchangé  
• Si $k < 0$ son sens est inversé  
• Si $k = 0$ on a : $0\overrightarrow{u} = \overrightarrow{0}$    

### 5) Propriété sur la bilinéarité 
$k(\overrightarrow{u}+\overrightarrow{v})=k\overrightarrow{u}+k\overrightarrow{v}$  
$(k+k')\overrightarrow{u}=k\overrightarrow{u}+k'\overrightarrow{u}$  

### 6) Colinéarité  
Soient $\overrightarrow{u}$ et $\overrightarrow{v}$ , deux vecteurs . Ils sont colinéaires entre eux si et seulement si $\overrightarrow{u}=k\overrightarrow{v}$.  

  
  $\overrightarrow{AB}=k\overrightarrow{CD} \iff (AB)//(CD)$  
  $\overrightarrow{AB}=k\overrightarrow{AC} \iff$ A,B et C sont alignés.  
  ### 7) Géométrie analytique  
 - Soient  $A(x_A;y_A)$ et $B(x_B;y_B)$ deux points du plan, alors $\overrightarrow{AB}(x_B-x_A;y_B-y_A)$.  
     
 - Soient les vecteurs $\overrightarrow{u}(x;y)$ et $\overrightarrow{v}(x';y')$, le déterminant des deux vecteurs est:  
  $det(\overrightarrow{u},\overrightarrow{v})=
  \begin{vmatrix} 
  x & x' \\ 
  y & y' 
  \end{vmatrix}
  = xy'-yx'$ 
      
 - $\overrightarrow{u}$ et $\overrightarrow{v}$ sont colinéaires $\iff det(\overrightarrow{u},\overrightarrow{v})=0$    
    
 - Dans un repère orthonormé, on peut calculer la norme $\begin{Vmatrix}\overrightarrow{u}\end{Vmatrix} $ et la distance $AB$. 
 $\begin{Vmatrix}\overrightarrow{u}\end{Vmatrix} = \sqrt{x^2+y^2}$  
 $AB = \sqrt{(x_B-x_A)^2+(y_B-y_A)^2}$.

 Soient $A(1;4)$ et $B(-5;2)$, calculer les coordonnées de $\overrightarrow{AB}$ et la norme $AB$.

## III) Orthogonalité  
### 1) Bilinéarité
Le produit scalaire est un forme bilinéaire symétrique :  
• symétrie : $\overrightarrow{u}.\overrightarrow{v} = \overrightarrow{v}.\overrightarrow{u}$  
• bilinéarité : pour tous réels $a$ et $b$ :  
$\overrightarrow{u} · (\overrightarrow{v} + \overrightarrow{w}) = \overrightarrow{u} ·\overrightarrow{v} +\overrightarrow{u} · \overrightarrow{w}$ et  
  $(a\overrightarrow{u}) · (b\overrightarrow{v}) = ab × (\overrightarrow{u} ·\overrightarrow{v})$   
  
  
  *Exemple:*  
  En utilisant les renseignements portés
sur la figure ci-contre, calculer le produit scalaire:  
$(\overrightarrow{AB}+\overrightarrow{AH}).\overrightarrow{AB}$  
  
    
  
 ### 2) Colinéarité et orthogonalité
  
  $\overrightarrow{u}.\overrightarrow{v}=||\overrightarrow{u}|| \times ||\overrightarrow{v}||  \iff \overrightarrow{u}$ et $\overrightarrow{v}$ sont de même sens  
  
  $\overrightarrow{u}.\overrightarrow{v}= - ||\overrightarrow{u}|| \times ||\overrightarrow{v}||  \iff \overrightarrow{u}$ et $\overrightarrow{v}$ sont de sens contraire  
  
  
  $\overrightarrow{u}.\overrightarrow{v}=0  \iff \overrightarrow{u}$ et $\overrightarrow{v}$ sont orthogonaux.  
  
  *remarque:*  
   $\overrightarrow{u}.\overrightarrow{u}=||\overrightarrow{u}|| \times ||\overrightarrow{u}|| = ||\overrightarrow{u}||^2$

## V) Formules
### 1) définition de la norme  
Le produit scalaire de deux vecteurs $\overrightarrow{u}$ et $\overrightarrow{v}$ est égal à :  
$$\overrightarrow{u}.\overrightarrow{v}=\frac{1}{2} \big( ||\overrightarrow{u}+\overrightarrow{v}||^2-||\overrightarrow{u}||^2-||\overrightarrow{v}||^2 \big)$$  
Cette définition mesure le défaut d’orthogonalité de deux vecteurs.
En effet si le triangle ABC est rectangle en A, alors $BC^2 − AB^2 − AC^2 = 0$   


*Exemple:*

### 2) Formule d'Al-Kashi  
Cette relation a pour but de déterminer une relation entre les trois longueurs d’un triangle, il s’agit de la généralisation du théorème de Pythagore.  
#### a) Définition  
Soit un triangle ABC. On a :  
$$BC^2 = AB^2 + AC^2 − 2 \times AB \times AC \times cos(\widehat{A})$$ 
  
Avec $a$, $b$ et $c$ les longueurs des côtés opposés respectivement à $A$, $B$ et $C$. On a : 
$$a^2 = b^2 + c^2 − 2bc \times  cos(\widehat{A})$$  
#### b) Démonstration
<div class="alert alert-block alert-danger">
    On cherche à démontrer la formule permettant d'exprimer $BC^2$. <br>  
    $\begin{array}\\   
    {BC}^2 & = & (\overrightarrow{BA}+\overrightarrow{AC})^2\\
     &=&(\overrightarrow{AC}-\overrightarrow{BA})^2\\
     & = & ||\overrightarrow{AC}||^2 - 2 \times \overrightarrow{AC}. \overrightarrow{BA} + ||\overrightarrow{BA}||^2\\
     & = & AC^2 + BA^2 - 2 \times AC \times BA \times cos(\widehat{A})\\
     \end{array}$  
</div>

## VI) Ensemble de points

### 1) Théorème de la médiane
Soient $A$ et $B$ deux points du plan et $I$ leur milieu. Pour tout point $M$, on a la relation:  
$$\overrightarrow{MA}.\overrightarrow{MB}= MI^2-\frac{1}{4}AB^2$$  


### 2) Démonstration    

<div class="alert alert-block alert-danger">
    On cherche à démontrer le théorème de la médiane. <br>  
    $\begin{array} \\  
    \overrightarrow{MA}.\overrightarrow{MB} & = & (\overrightarrow{MI}+\overrightarrow{IA}).(\overrightarrow{MI}+\overrightarrow{IB})\\
     &=& \overrightarrow{MI}^2 + \overrightarrow{MI}.\overrightarrow{IB} + \overrightarrow{IA}.\overrightarrow{MI} + \overrightarrow{IA}.\overrightarrow{IB}\\
     & = & \overrightarrow{MI}^2 + \overrightarrow{MI}.(\overrightarrow{IB} + \overrightarrow{IA}) + \overrightarrow{IA}.\overrightarrow{IB}\\
     & = &\overrightarrow{MI}^2 +\overrightarrow{IA}.\overrightarrow{IB}\\
     & = & \overrightarrow{MI}^2 +(-\frac{1}{2})\overrightarrow{AB}.\frac{1}{2}\overrightarrow{AB}\\  
     & = & \overrightarrow{MI}^2 -\frac{1}{4}\overrightarrow{AB}^2\\ 
     & = & MI^2 -\frac{1}{4}AB^2
     \end{array}$
</div>

### 3) Exemple    
Nous cherchons à définir l'ensemble des points $M$ tels que:  
$\overrightarrow{MA}.\overrightarrow{MB}=7$ avec $AB=6$.  
  
En utilisant le théorème de la médiane, on peut simplifier l'écriture:  
$\overrightarrow{MA}.\overrightarrow{MB}=7 \iff MI^2-\frac{1}{4}AB^2=7$  

Connaissant $AB=6$, on obtient:  
$\overrightarrow{MA}.\overrightarrow{MB}=7 \iff MI^2-\frac{1}{4}36=7$  
$\overrightarrow{MA}.\overrightarrow{MB}=7 \iff MI^2-9=7$  
$\overrightarrow{MA}.\overrightarrow{MB}=7 \iff MI^2=...$  
$\overrightarrow{MA}.\overrightarrow{MB}=7 \iff MI=...$
